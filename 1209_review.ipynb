### 합계
* 열과 행 방향이 있어서 axis 필요
* df.sum()  # 열기준 (axis=0) 열을 남기고 행을 축소 = 행 내부 합계
* df.sum(axis=1) # 행방향 합계, 열 축소 = 열들 합
* df.mean() #행 인덱스가  = 열 내부 헹들간 합계
* df.mean(axis=1) #행 인덱스가  = 행 내부 헹들간 합계

### `apply` 변환
* sum : 합계, mean : 평균 ...
* 임의 정의된 함수를 행/열 단위로 적용하고 싶어!
* df.apply(함수, axis=축)
* 1. 함수 -> 이미 정의된 함수
* print, ... apply => '인자'로 값 한 개를 받는 함수
* def fun1(x): # axis=0 => 열 뱡향으로 원소들을 반복 처리
*	return .... => 특정한 값을 가지는 함수
* 익명함수 lambda (인자들) : (인자들을 사용한 표현식)
* lambda p : p => (return)
* def fun(p):
*     return p
* lambda x: x. ...
* 2.
* axis = 0 : 열 (default)
* df.apply(...) # 모든 열에 대해서 처리
* df['열 이름'].apply(...) # 특정한 열에 대해서 처리
* axis = 1 : 행
* df.apply(..., axis=1)
* lambda row(행) : row.컬럼들을 호출해서 통계적 작업
